```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'scraped_data/emb/COMPRESSORS-2022-04-23-15-30-03.csv')
df.head()
```

```python
# Standardize columns
df.columns = df.columns.str.replace(" ", "").str.lower()
df.head()
```

```python
# rename some columns
rename_columns = {'frequency/rotation': 'frequency_rotation',
                  'powersupply': 'power_supply',
                  'standardcondition': 'standard_conditions',
                  'testedconditions': 'tested_conditions',
                  'motortype': 'motor_type',
                  'startingtorque': 'starting_torque'}

df = df.rename(columns=rename_columns)
df.head()
```

```python
# explore unique values
for col in df.columns:
    print(col, df[col].unique(), sep='\n', end='\n\n')
```

```python
# Model soft serch
df.loc[df.model.isin([model for model in df.model if 'EMY62' in model])]
```
